**TASK 2A: AUSTRALIAN COLOURING MAP (CSP)**

```python


# Regions
regions = ["WA", "NT", "SA", "Q", "NSW", "V", "T"]

# Neighbors (constraints: no adjacent regions share color)
neighbors = {
    "WA": ["NT", "SA"],
    "NT": ["WA", "SA", "Q"],
    "SA": ["WA", "NT", "Q", "NSW", "V"],
    "Q": ["NT", "SA", "NSW"],
    "NSW": ["Q", "SA", "V"],
    "V": ["SA", "NSW"],
    "T": []  # Tasmania has no neighbors
}

# Colors
colors = ["Red", "Green", "Blue"]

# Assignment dictionary
assignment = {}

# Check if safe
def is_safe(region, color):
    for neighbor in neighbors[region]:
        if neighbor in assignment and assignment[neighbor] == color:
            return False
    return True

# Backtracking function
def backtrack():
    if len(assignment) == len(regions):
        return True

    # Select unassigned region
    for region in regions:
        if region not in assignment:
            for color in colors:
                if is_safe(region, color):
                    assignment[region] = color

                    if backtrack():
                        return True


                    del assignment[region]
            return False


backtrack()


print("Australia Map Coloring Solution:")
for region in assignment:
    print(region, "→", assignment[region])
```

**TASK 2B: NAIROBI SUB-COUNTIES COLOURING(CSP)**

```python


counties = [
    "Westlands", "Dagoretti North", "Dagoretti South",
    "Langata", "Kibra", "Roysambu", "Kasarani",
    "Ruaraka", "Embakasi South", "Embakasi North",
    "Embakasi Central", "Embakasi East", "Embakasi West",
    "Makadara", "Kamukunji", "Starehe", "Mathare"
]


neighbors_nairobi = {
    "Westlands": ["Dagoretti North", "Starehe"],
    "Dagoretti North": ["Westlands", "Dagoretti South"],
    "Dagoretti South": ["Dagoretti North", "Langata"],
    "Langata": ["Dagoretti South", "Kibra"],
    "Kibra": ["Langata", "Starehe"],
    "Roysambu": ["Kasarani"],
    "Kasarani": ["Roysambu", "Ruaraka"],
    "Ruaraka": ["Kasarani", "Embakasi North"],
    "Embakasi North": ["Ruaraka", "Embakasi Central"],
    "Embakasi Central": ["Embakasi North", "Embakasi East"],
    "Embakasi East": ["Embakasi Central", "Embakasi West"],
    "Embakasi West": ["Embakasi East", "Makadara"],
    "Makadara": ["Embakasi West", "Kamukunji"],
    "Kamukunji": ["Makadara", "Starehe"],
    "Starehe": ["Westlands", "Kamukunji", "Kibra"],
    "Mathare": ["Ruaraka"],
}


colors = ["Red", "Green", "Blue"]

assignment_nairobi = {}

def is_safe_nairobi(region, color):
    for neighbor in neighbors_nairobi.get(region, []):
        if neighbor in assignment_nairobi and assignment_nairobi[neighbor] == color:
            return False
    return True

def backtrack_nairobi():
    if len(assignment_nairobi) == len(counties):
        return True

    for region in counties:
        if region not in assignment_nairobi:
            for color in colors:
                if is_safe_nairobi(region, color):
                    assignment_nairobi[region] = color

                    if backtrack_nairobi():
                        return True

                    del assignment_nairobi[region]
            return False


backtrack_nairobi()

print("\nNairobi Sub-Counties Coloring:")
for region in assignment_nairobi:
    print(region, "→", assignment_nairobi[region])
```
